````python
def get_business_entity(provider_to_be_checked, location_to_be_checked):
        logging.info('Starting business entity mapping')
      # Load the providers and locations JSON data.
        # logging.info(location_to_be_checked)
        with open("/mnt/nfs_fbs_client/charges/sutterInpatientCharges/businessEntityJson/updated_entity_id_providers.json") as providers:
            providers_json = json.load(providers)
        with open("/mnt/nfs_fbs_client/charges/sutterInpatientCharges/businessEntityJson/updated_entity_id_locations.json") as locations:
            locations_json = json.load(locations)

          # Check if the provider is present in the providers JSON.
        for provider in providers_json:
            
            if provider_to_be_checked == provider["Provider"]:
                logging.info(f'Provider match found in business entity mapping, id: {provider["CareCloud business entity"]}')
                return {"id":provider["CareCloud business entity"],"name":provider["Provider"]}

          # Check if the location is present in the locations JSON.
        for location in locations_json:
            if location_to_be_checked == location["Location"]:
                logging.info(f'Provider  match not found, Location match found in business entity mapping ,id:{location["CareCloud business entity"]}')
                return {"id":location["CareCloud business entity"],"location":location["Location"]}

          # The provider or location is not present.
        return (
                    logging.warning(f'Provider match found in business entity mapping, id: {provider["CareCloud business entity"]}'),
                    {"error": {"provider_to_be_checked": provider_to_be_checked, "location_to_be_checked": location_to_be_checked}}
                )

````

````python
#error files using timestamp as key 
import os
import json

def combine_json_files(directory, output_file):
    combined_data = {}
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        if os.path.isfile(file_path) and filename.endswith('.json'):
            with open(file_path) as f:
                data = json.load(f)
            timestamp = filename.split('_')[1].split('.')[0]  # Extract the timestamp from the filename
            if timestamp in combined_data:
                combined_data[timestamp].update(data)  # Update the existing dictionary with the new data
            else:
                combined_data[timestamp] = data

    with open(output_file, 'w') as f:
        print(len(combined_data["charges"]))
        json.dump(combined_data, f, indent=2)

    logging.info("Combined JSON files successfully.")
    

````
